### Introdução à Linguagem Python e suas Estruturas Básicas

```python
# 1. Introdução


print("Olá mundo!")

var_1 = 2.36
var_2 = 12

print(f'Valor da variavel var_1 = {var_1}')
```

```python
# 2. Variáveis e Tipos de Dados

inteiro = 10
flutuante = 10.5
texto = 'Ola mundo'
booleano = True
```

```python
# 3. Estruturas Condicionais

numero = 7

if numero % 2 == 0:
  print(f'O numero {numero} é par')
elif numero > 10:
  print(f'O numero {numero} é maior que 10')
else
  print(f'O numero {numero} é impar')
```

```python
# 4. Estruturas de Repetição

print("Contando de 1 a 5 usando o for:")

for i in range (1,6):
  print(i)

print('\nContando de 0 a 9 usando o while:')
contador = 0

while contador < 10:
    print(contador)
    contador += 1
```

```python
# 5. Listas

lista = [1,2,3,4,5]
print("Lista original", lista)
lista.append(6)
print("Lista modificada", lista)

print("Posicao de numero 2 da lista: ", lista[1])

lista.pop(2)
print("Lista modificada apos o POP", lista)

lista.insert(2, 3)
print("Lista modificada apos o INSERT", lista)


print("Tamanho da lista: ", len(lista))

print("Somar os itens da lista: ", sum(lista))

print(55 in lista)

# Concatenar listas

lista.extend([7,8,9])
print("Lista apos EXTEND: ", lista)

a,b = [1,2]

print(a)
print(b)
```

```python
# 6. Dicionarios

dicionario = {'nome': "Alice", 'idade': 25, 'cidade': 'Sao Paulo'}

print(dicionario["nome"])

dicionario['idade'] = 26
print("dicionario apos update: ", dicionario)
```

```python
# 7. Funções

def saudacao(nome):
  print(f"Ola {nome}!")

saudacao("Alice")

def my_print(message = "Hello world!"):
  return message

saudacao(my_print("Joao"))
```

```python
# 8. Erros / Exceções

try:
  print(0/0)
except:
  print("Divisao por 0")

  x = 1e-100
  print(x ** 0)
  print(x ** 0.99)
  print(x ** 0.00001)
  print(x ** 0.00000001)
```

```python
# 9. Valores Nulos

u = None
print(u is None)

print(u == None) # nao é o correto
```

```python
# 10. Classe e Objetos

class Pessoa:
  def __init__(self,nome,idade):
    self.nome = nome
    self.idade = idade

  def apresentar(self):
   return print(f"Ola, meu nome é {self.nome} e tenho {self.idade} anos.")

pessoa = Pessoa("Alice", 25)
print(pessoa.apresentar())
```

```python
# Exercicios

numeros = []
class categoria:
  def __init__(self,numeros) :
    self.numeros = numeros
    self.pares = []
    self.impares = []

  def categorizar(self):
      for i in self.numeros:
        if i % 2 == 0:
          self.pares.append(i)
        else:
          self.impares.append(i)

  def imprimir(self):
       print(f"Pares: ({self.pares}), Soma:  {sum(self.pares)},\n Impares: ({self.impares}), Soma:  {sum(self.impares)}")


lista1 = categoria([12, 2, 3, 4, 5, 6, 7, 8, 9])
lista1.categorizar()
lista1.imprimir()


```
